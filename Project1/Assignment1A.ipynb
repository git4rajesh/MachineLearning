```
import numpy as np
from urllib.request import urlopen
import cv2
from matplotlib import pyplot as plt

# METHOD #1: OpenCV, NumPy, and urllib
def url_to_image(url):
	# download the image, convert it to a NumPy array, and then read
	# it into OpenCV format
	resp = urlopen(url)
	image = np.asarray(bytearray(resp.read()), dtype="uint8")
	image = cv2.imdecode(image, cv2.IMREAD_COLOR)
 
	# return the image
	return image
```

```
image = url_to_image("https://fontsarena-cd5e.kxcdn.com/wp-content/uploads/2019/04/helvetica-now-font-400x364.png")
```

# Horizontal Edge Detector

```
# Creating a kernel of 3 * 3 matrix which is a transpose of vertical detector kernel used in example
kernel = np.array([[-1,-1,-1],[0,0,0],[1,1,1]])

# Multiplying the kernel by a factor of 1 to get the Horizontal Edge Detector Kernel
hor_kernel = 1 * (kernel)

# Convolves an image with the kernel. The function applies an arbitrary linear filter to an image
dst = cv2.filter2D(image,-1, hor_kernel)

# Display an image in a window
cv2_imshow(dst)
```

```
# Using a Sobel Horizontal Kernel for Horizontal Identification
hor_kernel = np.array([[1, 2, 1],[0, 0, 0],[-1, -2,-1]])
dst = cv2.filter2D(image,-1, hor_kernel)
cv2_imshow(dst)
```

```
# Modfiying the Sobel kernel by mutiplying with a factor of 1.5
kernel = np.array([[1, 2, 1],[0, 0, 0],[-1, -2,-1]])

hor_kernel = 3/2 * kernel
dst = cv2.filter2D(image,-1, hor_kernel)
cv2_imshow(dst)
```

# Blur Kernel

```
# Blur Kernel 

# Creating a 3 * 3 ones matrix 
kernel = np.ones([3, 3])

# Dividing it by a factor of 9 so that kernel is weak.
# Box Blur where Each pixel is computed as the average of the surrounding pixels
blur_kernel = (1/9) * kernel

# Convolution using Blur kernel
dst = cv2.filter2D(image,-1,blur_kernel)

# Show the Blurred Image
cv2_imshow(dst)
```

# SHARPEN KERNEL

```
# In this kernel, the pixel is boosted when the neighbor pixels are different.
# The details of the image is highlighted by using a high-pass filter
kernel = np.float32([(0, -1, 0), (-1, 5, -1), (0, -1, 0)])

sharpen_kernel = 5/4 * kernel
                          
dst = cv2.filter2D(image,-1, sharpen_kernel)
cv2_imshow(dst)
```

# IDENTITY KERNEL

```
# Identity Kernel is that kernel which always returns the same output image as its input image
kernel = np.float32([(0, 0, 0), (0, 1, 0), (0, 0, 0)])

identity_kernel = kernel
                          
dst = cv2.filter2D(image,-1, identity_kernel)
cv2_imshow(dst)
```
